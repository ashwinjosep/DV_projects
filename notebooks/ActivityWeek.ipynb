```python
def get_days(days0):
    for i in range(len(days0)):
        if days0[i] == 0:
            days0[i] = "Monday"
        elif days0[i] == 1:
            days0[i] = "Tuesday"
        elif days0[i] == 2:
            days0[i] = "Wednesday"
        elif days0[i] == 3:
            days0[i] = "Thursday"
        elif days0[i] == 4:
            days0[i] = "Friday"
        elif days0[i] == 5:
            days0[i] = "Saturday"
        else:
            days0[i] = "Sunday"
    return days0
```

```python
from collections import Counter
```

```python
days = []
for checkin in checkins:
    each_days = []
    each = checkin.split(', ')
    each = [ck.split(' ')[0] for ck in each]
    for dt in each:
        each_days.append(date(int(dt.split('-')[0]), int(dt.split('-')[1]), int(dt.split('-')[2])).weekday())
    each_days1 = get_days(each_days)
    days.append(dict(Counter(each_days1)))
```
